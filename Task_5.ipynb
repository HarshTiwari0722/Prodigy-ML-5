```python
!pip install mediapipe opencv-python tensorflow scikit-learn matplotlib seaborn
!pip install pillow numpy pandas
!pip install mediapipe opencv-python scikit-learn seaborn
import cv2
import mediapipe as mp
import numpy as np
import pandas as pd
import tensorflow as tf
print("All imports successful!")

import subprocess
import sys

def install_requirements():
    """Install all required packages"""
    packages = [
        'mediapipe',
        'opencv-python',
        'scikit-learn',
        'seaborn'
    ]

    for package in packages:
        print(f"Installing {package}...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

    print("All packages installed successfully!")

install_requirements()

# ============================================================================
# STEP 2: Import Libraries
# ============================================================================
import cv2
import mediapipe as mp
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
from google.colab import files
import io
from PIL import Image
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported successfully!")

# ============================================================================
# STEP 3: Initialize MediaPipe Hand Detection
# ============================================================================
class HandGestureRecognizer:
    def __init__(self):
        self.mp_hands = mp.solutions.hands
        self.hands = self.mp_hands.Hands(
            static_image_mode=False,
            max_num_hands=2,
            min_detection_confidence=0.7,
            min_tracking_confidence=0.5
        )
        self.mp_drawing = mp.solutions.drawing_utils
        self.model = None
        self.label_encoder = None

    def extract_landmarks(self, image):
        """Extract hand landmarks from image"""
        rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        results = self.hands.process(rgb_image)

        landmarks_list = []
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                landmarks = []
                for lm in hand_landmarks.landmark:
                    landmarks.extend([lm.x, lm.y, lm.z])
                landmarks_list.append(landmarks)

        return landmarks_list, results

    def draw_landmarks(self, image, results):
        """Draw hand landmarks on image"""
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                self.mp_drawing.draw_landmarks(
                    image, hand_landmarks, self.mp_hands.HAND_CONNECTIONS)
        return image

# ============================================================================
# STEP 4: Data Collection Function
# ============================================================================
def collect_gesture_data():
    """
    Collect hand gesture data for training
    This function helps you create a dataset by capturing gestures
    """
    recognizer = HandGestureRecognizer()

    # Define gesture classes
    gestures = ['thumbs_up', 'thumbs_down', 'peace', 'fist', 'open_palm', 'pointing']

    print("Gesture Data Collection")
    print("======================")
    print("Available gestures:", gestures)
    print("Instructions:")
    print("1. Choose a gesture from the list above")
    print("2. Upload images of that gesture")
    print("3. Repeat for all gestures you want to recognize")

    # For demonstration, we'll create synthetic data
    # In real implementation, you would collect actual images

    data = []
    labels = []

    # Generate sample data (replace with actual image collection)
    np.random.seed(42)
    for gesture in gestures:
        # Generate 100 samples per gesture (synthetic data for demo)
        for _ in range(100):
            # Create random landmarks (21 points * 3 coordinates = 63 features)
            landmarks = np.random.rand(63)
            data.append(landmarks)
            labels.append(gesture)

    return np.array(data), np.array(labels)

# ============================================================================
# STEP 5: Data Preprocessing
# ============================================================================
def preprocess_data(data, labels):
    """Preprocess the collected data"""

    # Encode labels
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)

    # Convert to categorical
    num_classes = len(np.unique(encoded_labels))
    categorical_labels = keras.utils.to_categorical(encoded_labels, num_classes)

    # Normalize data
    data = data.astype('float32')
    data = (data - np.mean(data, axis=0)) / (np.std(data, axis=0) + 1e-8)

    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        data, categorical_labels, test_size=0.2, random_state=42, stratify=encoded_labels
    )

    return X_train, X_test, y_train, y_test, label_encoder, num_classes

# ============================================================================
# STEP 6: Model Architecture
# ============================================================================
def create_gesture_model(input_shape, num_classes):
    """Create deep learning model for gesture recognition"""

    model = keras.Sequential([
        keras.layers.Input(shape=input_shape),

        # Dense layers with dropout for regularization
        keras.layers.Dense(128, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),

        keras.layers.Dense(256, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.4),

        keras.layers.Dense(128, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.3),

        keras.layers.Dense(64, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.2),

        # Output layer
        keras.layers.Dense(num_classes, activation='softmax')
    ])

    # Compile model
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# ============================================================================
# STEP 7: Training Function
# ============================================================================
def train_model(X_train, X_test, y_train, y_test, num_classes):
    """Train the gesture recognition model"""

    # Create model
    model = create_gesture_model((X_train.shape[1],), num_classes)

    print("Model Architecture:")
    model.summary()

    # Callbacks
    early_stopping = keras.callbacks.EarlyStopping(
        monitor='val_accuracy', patience=10, restore_best_weights=True
    )

    reduce_lr = keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', factor=0.5, patience=5, min_lr=1e-7
    )

    # Train model
    print("\nTraining the model...")
    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=100,
        batch_size=32,
        callbacks=[early_stopping, reduce_lr],
        verbose=1
    )

    return model, history

# ============================================================================
# STEP 8: Model Evaluation
# ============================================================================
def evaluate_model(model, X_test, y_test, label_encoder, history):
    """Evaluate the trained model"""

    # Predictions
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)

    # Classification report
    class_names = label_encoder.classes_
    print("Classification Report:")
    print(classification_report(y_true_classes, y_pred_classes, target_names=class_names))

    # Confusion Matrix
    plt.figure(figsize=(12, 5))

    # Plot 1: Confusion Matrix
    plt.subplot(1, 2, 1)
    cm = confusion_matrix(y_true_classes, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')

    # Plot 2: Training History
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return y_pred, y_pred_classes

# ============================================================================
# STEP 9: Real-time Gesture Recognition
# ============================================================================
def real_time_gesture_recognition(model, label_encoder):
    """Real-time gesture recognition using webcam"""

    recognizer = HandGestureRecognizer()
    recognizer.model = model
    recognizer.label_encoder = label_encoder

    print("Real-time Gesture Recognition")
    print("=============================")
    print("Instructions:")
    print("1. Allow camera access when prompted")
    print("2. Show your hand gestures to the camera")
    print("3. Press 'q' to quit (if running locally)")

    # For Colab, we'll use uploaded images instead of webcam
    print("\nFor Google Colab: Upload an image to test gesture recognition")

    def predict_gesture_from_image(image):
        """Predict gesture from uploaded image"""
        landmarks_list, results = recognizer.extract_landmarks(image)

        if landmarks_list:
            # Use first detected hand
            landmarks = np.array(landmarks_list[0]).reshape(1, -1)

            # Normalize (same as training data)
            landmarks = (landmarks - np.mean(landmarks)) / (np.std(landmarks) + 1e-8)

            # Predict
            prediction = model.predict(landmarks, verbose=0)
            predicted_class = np.argmax(prediction)
            confidence = np.max(prediction)
            gesture_name = label_encoder.inverse_transform([predicted_class])[0]

            return gesture_name, confidence, results

        return None, 0, results

    return predict_gesture_from_image

# ============================================================================
# STEP 10: Image Upload and Testing Function
# ============================================================================
def test_with_uploaded_image(predict_function):
    """Test gesture recognition with uploaded images"""

    print("Upload an image to test gesture recognition:")
    uploaded = files.upload()

    for filename in uploaded.keys():
        print(f"\nProcessing {filename}...")

        # Load image
        image_data = uploaded[filename]
        image = Image.open(io.BytesIO(image_data))
        image_np = np.array(image)

        # Convert RGB to BGR for OpenCV
        if len(image_np.shape) == 3:
            image_cv = cv2.cvtColor(image_np, cv2.COLOR_RGB2BGR)
        else:
            image_cv = image_np

        # Predict gesture
        gesture, confidence, results = predict_function(image_cv)

        # Draw landmarks
        output_image = image_cv.copy()
        recognizer = HandGestureRecognizer()
        output_image = recognizer.draw_landmarks(output_image, results)

        # Display results
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(image_cv, cv2.COLOR_BGR2RGB))
        plt.title('Original Image')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
        if gesture:
            plt.title(f'Detected: {gesture}\nConfidence: {confidence:.2f}')
        else:
            plt.title('No hand detected')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

# ============================================================================
# STEP 11: Save and Load Model
# ============================================================================
def save_model(model, label_encoder, filename='gesture_model'):
    """Save the trained model and label encoder"""

    # Save model
    model.save(f'{filename}.h5')

    # Save label encoder
    with open(f'{filename}_encoder.pkl', 'wb') as f:
        pickle.dump(label_encoder, f)

    print(f"Model saved as {filename}.h5")
    print(f"Label encoder saved as {filename}_encoder.pkl")

def load_model(filename='gesture_model'):
    """Load the trained model and label encoder"""

    # Load model
    model = keras.models.load_model(f'{filename}.h5')

    # Load label encoder
    with open(f'{filename}_encoder.pkl', 'rb') as f:
        label_encoder = pickle.load(f)

    print(f"Model loaded from {filename}.h5")
    return model, label_encoder

# ============================================================================
# STEP 12: Main Execution Function
# ============================================================================
def main():
    """Main function to run the complete gesture recognition system"""

    print("Hand Gesture Recognition System")
    print("===============================")

    # Step 1: Collect/Generate Data
    print("\n1. Collecting gesture data...")
    data, labels = collect_gesture_data()
    print(f"Data shape: {data.shape}")
    print(f"Labels shape: {labels.shape}")
    print(f"Unique gestures: {np.unique(labels)}")

    # Step 2: Preprocess Data
    print("\n2. Preprocessing data...")
    X_train, X_test, y_train, y_test, label_encoder, num_classes = preprocess_data(data, labels)
    print(f"Training data shape: {X_train.shape}")
    print(f"Test data shape: {X_test.shape}")
    print(f"Number of classes: {num_classes}")

    # Step 3: Train Model
    print("\n3. Training model...")
    model, history = train_model(X_train, X_test, y_train, y_test, num_classes)

    # Step 4: Evaluate Model
    print("\n4. Evaluating model...")
    y_pred, y_pred_classes = evaluate_model(model, X_test, y_test, label_encoder, history)

    # Step 5: Save Model
    print("\n5. Saving model...")
    save_model(model, label_encoder)

    # Step 6: Set up real-time recognition
    print("\n6. Setting up real-time recognition...")
    predict_function = real_time_gesture_recognition(model, label_encoder)

    # Step 7: Test with uploaded image
    print("\n7. Ready for testing!")
    print("You can now upload images to test gesture recognition.")

    return model, label_encoder, predict_function

# ============================================================================
# STEP 13: Run the Complete System
# ============================================================================
if __name__ == "__main__":
    # Run the main function
    model, label_encoder, predict_function = main()

    # Test with uploaded images
    print("\n" + "="*50)
    print("TESTING PHASE")
    print("="*50)
    test_with_uploaded_image(predict_function)
```
